# 🐍 Funciones Built-in de Python
### Ejemplos prácticos y claros

---
En este notebook exploraremos 6 funciones integradas muy útiles de Python:
`dir()`, `help()`, `eval()`, `reversed()`, `filter()` y `map()`.

---
## 1. `dir()` — Inspeccionar objetos

Te sirve para ver **"qué hay dentro"** de un objeto: sus atributos y métodos disponibles.

```python
lista = [1, 2, 3]

# Esto muestra todos los métodos que puedes usar con una lista (append, pop, etc.)
print(dir(lista))
```

```python
# Tip: filtra solo los métodos públicos (sin guiones bajos)
metodos_publicos = [m for m in dir(lista) if not m.startswith('_')]
print(metodos_publicos)
```

---
## 2. `help()` — Manual interactivo

Te da la **documentación oficial** directamente en tu consola.

```python
# Esto muestra el manual explicando qué hace la función 'print'
help(print)
```

```python
# También puedes consultarlo sobre métodos específicos de una clase
help(list.append)
```

---
## 3. `eval()` — Ejecutar texto como código

⚠️ **¡Úsala con precaución!** Evalúa una cadena de texto como si fuera código Python y devuelve el resultado.

```python
operacion = "2 + 5 * 10"

# Evalúa el texto y devuelve el resultado matemático: 52
resultado = eval(operacion)
print(resultado)  # 52
```

```python
# Otro ejemplo: la calculadora dinámica
expresion = input("Escribe una operación matemática: ")
print("Resultado:", eval(expresion))
```

---
## 4. `reversed()` — Invertir secuencias

Devuelve un **iterador** que recorre la secuencia de atrás hacia adelante.

```python
numeros = [1, 2, 3, 4]

for n in reversed(numeros):
    print(n)  # Imprimirá: 4, 3, 2, 1
```

```python
# También funciona con strings
palabra = "Python"
al_reves = "".join(reversed(palabra))
print(al_reves)  # nohtyP
```

---
## 5. `filter()` — Filtrar elementos

Crea una nueva colección **solo con los elementos que cumplen una condición**.

> 💡 `lambda` es una forma de crear funciones rápidas y anónimas en una sola línea. Aquí le dice a `filter()` qué condición usar.

```python
numeros = [1, 2, 3, 4, 5, 6]

# Filtra solo los números pares
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)  # [2, 4, 6]
```

```python
# Sin lambda: usando una función definida
def es_par(x):
    return x % 2 == 0

pares2 = list(filter(es_par, numeros))
print(pares2)  # [2, 4, 6]  ← mismo resultado
```

---
## 6. `map()` — Transformar elementos

**Aplica una operación a cada elemento** de una lista y devuelve los resultados.

> 💡 Igual que en `filter()`, `lambda` define la transformación que se aplica a cada elemento.

```python
numeros = [1, 2, 3, 4]

# Eleva cada número al cuadrado
cuadrados = list(map(lambda x: x**2, numeros))
print(cuadrados)  # [1, 4, 9, 16]
```

```python
# Sin lambda: usando una función definida
def al_cuadrado(x):
    return x ** 2

cuadrados2 = list(map(al_cuadrado, numeros))
print(cuadrados2)  # [1, 4, 9, 16]  ← mismo resultado
```

## 📋 Resumen rápido

| Función | ¿Qué hace? | Ejemplo rápido |
|---|---|---|
| `dir()` | Lista métodos y atributos de un objeto | `dir([1,2,3])` |
| `help()` | Muestra la documentación oficial | `help(print)` |
| `eval()` | Ejecuta texto como código Python | `eval("2+2")` → 4 |
| `reversed()` | Itera una secuencia al revés | `list(reversed([1,2,3]))` |
| `filter()` | Filtra elementos por condición | `filter(lambda x: x>2, lista)` |
| `map()` | Transforma cada elemento | `map(lambda x: x*2, lista)` |
